```python
import pandas as pd
```

```python
from bokeh.io import output_notebook, show
output_notebook()
```

```python
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource, HoverTool
```

```python
uk_data = pd.read_excel("/home/baali/projects/ai-ml-workshop/data/uk-population-age-2018.xlsx")
uk_data = uk_data.drop([0, 2])
uk_data = uk_data.loc[:93]
uk_data.columns = ["age", "2018-m", "2008-m", "2018-f", "2008-f"]
uk_data["total"] = uk_data["2018-m"]+uk_data["2018-f"]
uk_source = ColumnDataSource(uk_data)

p = figure(x_range=(0, 90), y_range=(0, max(uk_data['total'][1:])+50000), plot_height=400, title="UK Age distribution of Population 2018",toolbar_location=None, tools="")
p.vbar(x='age', top='total', width=0.9, source=uk_source)
p.add_tools(HoverTool(tooltips=[("Population", "@total"), ("Age", "@age")]))
show(p)
```

```python
fr_data = pd.read_csv("/home/baali/projects/ai-ml-workshop/data/france-population.csv")
fr_source = ColumnDataSource(fr_data)

p = figure(x_range=(0, 102), y_range=(0, max(fr_data['total'])+100000), plot_height=400, title="France age distribution of Population 2019",toolbar_location=None, tools="")
p.vbar(x='age', top='total', width=0.9, source=fr_source)
p.add_tools(HoverTool(tooltips=[("Population", "@total"), ("Age", "@age")]))
show(p)
```

```python
data = pd.read_excel("http://www.censusindia.gov.in/2011census/C-series/c-13/DDW-0000C-13.xls", header=None)
data = data.loc[7:]
columns_to_remove = data.columns[[0,1,2]]
data.drop(columns_to_remove, axis=1, inplace=True)
indiadf = data.loc[:108]
indiadf.columns = [ "place" , "age" , "total" , "males" , "females" , "rural_total" , "rural_males" , "rural_females" , "urb_total" , "urb_males" , "urb_females" ]
source = ColumnDataSource(indiadf)

p = figure(x_range=(0, 101), y_range=(0, max(indiadf['total'][1:])+5000000), plot_height=400, title="Age distribution of Population",toolbar_location=None, tools="")
p.vbar(x="age", top='total', width=1, source=source, fill_color="#b3de69")
p.add_tools(HoverTool(tooltips=[("Population", "@total"), ("Age", "@age")]))
p.vbar(x="age", top='rural_total', width=1, source=source, fill_color="#ff0000")
p.vbar(x="age", top='urb_total', width=1, source=source)
show(p)
```
